```
import numpy as np
import plotly.graph_objects as go

# quick random 3D scatter
rng = np.random.default_rng(42)
pts = rng.standard_normal((500, 3)) * [1.0, 1.0, 0.3]

fig = go.Figure(data=[go.Scatter3d(
    x=pts[:,0], y=pts[:,1], z=pts[:,2], mode="markers", marker=dict(size=2)
)])
fig.update_layout(height=500, margin=dict(l=0,r=0,b=0,t=30), title="Random 3D points")
fig.show()
```
